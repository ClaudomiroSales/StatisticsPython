In this notebook, we'll explore different ways to create histograms using Python. We'll use the following libraries:


*   matplotlib
*   pandas
*   seaborn
*   plotly





```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# Number of samples

```python
n_samples = 10000
n_bins = 30 #default = 30
```

# Using Matplotlib

Here's a simple example of creating a histogram using `matplotlib`.

```python
# Generate random data
data = np.random.randn(n_samples)

# Create histogram
plt.hist(data, bins=n_bins, edgecolor='black')

# Add titles and labels
plt.title('Histogram using Matplotlib')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

# Using Pandas

If you have data in a pandas DataFrame, you can easily plot histograms using the `plot` method.

```python
# Create a DataFrame with random data
df = pd.DataFrame({'data': np.random.randn(n_samples)})

# Plot histogram
df['data'].plot(kind='hist', bins=n_bins, edgecolor='black', title='Histogram using Pandas')

# Show plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

```

# Using Seaborn

Seaborn is another powerful visualization library based on `matplotlib` that provides a high-level interface for drawing attractive statistical graphics.

```python
# Generate random data
data = np.random.randn(n_samples)

# Create histogram
sns.histplot(data, bins=n_bins, kde=True)

# Add titles and labels
plt.title('Histogram using Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

# Pandas and Seaborn

```python
# Generate random data and create a DataFrame
data = np.random.randn(n_samples)
df = pd.DataFrame({'data': data})

# Plot histogram using Seaborn
sns.histplot(df['data'], bins=n_bins, kde=True)

# Add titles and labels
plt.title('Histogram using Pandas and Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

# Using Plotly

Plotly is another popular library for interactive plots.

```python
import plotly.express as px

# Generate random data
data = np.random.randn(n_samples)

# Create histogram
fig = px.histogram(data, nbins=n_bins, title='Histogram using Plotly')

# Show plot
fig.show()
```

These examples should give you a good starting point for creating histograms with different libraries in Python. Each library has its own strengths, and you can choose the one that best fits your needs.
